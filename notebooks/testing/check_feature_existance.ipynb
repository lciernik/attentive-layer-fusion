## Notebook to check pre-extracted feature existance
This notebook checks whether the pre-extracted features for different datasets/models exist in the expected locations.

```python
import pandas as pd
import sys
sys.path.append('..')
sys.path.append('../..')

from constants import model_config_file, BASE_PATH_PROJECT, ds_list_file
from helper import load_all_datasetnames_n_info
from src.utils.utils import check_feature_existence
```

```python
ds_list, ds_info = load_all_datasetnames_n_info(ds_list_file)
```

```python
model_config = pd.read_json(model_config_file).T
```

```python
root_feature_dir = BASE_PATH_PROJECT / "features"
root_feature_dir
```

```python
from loguru import logger
logger.remove()

for ds in ds_list:
    print(f"{ds=}")
    for mid in model_config.index.tolist():
        curr_dir  = root_feature_dir / ds / mid 
        exists = check_feature_existence(
            curr_dir,
            module_names=model_config.loc[mid, 'module_names']
        ) 
        if not exists:
            print(f">> {mid=} features and targets are missing")
            print()
    print()
        
```
